## JSON PARSERS

### 1. Get index values

```python
def get_index(json_obj):
    return [data_obj['index'] for data_obj in json_obj['data']]
```

### 2. Get date values

```python
def get_date(json_obj, index_lst):
    date_lst = []
    for i in index_lst:
        date_lst.append(json_obj['data'][i]['items']['Tarih'])
    return date_lst
```

### 3. Week number of year values

```python
def get_weeokOfyear(json_obj,index_lst):
    weeokOfyear_lst = []
    for i in index_lst:
        weeokOfyear_lst.append(json_obj['data'][i]['items']['YEARWEEK'])
    return weeokOfyear_lst
```

### 4. Get date values

```python
def get_value(json_obj,index_lst):
    value_lst = []
    for i in index_lst:
        value_lst.append(json_obj['data'][i]['items']['TP_KKHARTUT_KT1'])
    return value_lst
```

```python
index_lst = get_index(kkhartut_data)
```

```python
dates = get_date(kkhartut_data, index_lst)
weeks = get_weeokOfyear(kkhartut_data, index_lst)
values = get_value(kkhartut_data, index_lst)
```

## CREATE DATAFRAME

```python
data = pd.DataFrame.from_dict({'date':dates, 'weeks':weeks, 'values':values})
```

### 6. Seasonal Decompose with Period 12

### 6.1 Multiplicative Seasonal Decompose with Period 12

```python
from statsmodels.tsa.seasonal import seasonal_decompose

#data.set_index('date', inplace=True)

analysis = data[['values']].copy()
analysis.sort_index(inplace=True)

decompose_result_mult = seasonal_decompose(analysis, model="multiplicative", extrapolate_trend='freq', period=12)

trend = decompose_result_mult.trend
seasonal = decompose_result_mult.seasonal
residual = decompose_result_mult.resid

fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(15,8))
decompose_result_mult.trend.plot(ax=ax1)
ax1.set(ylabel='Trend')
decompose_result_mult.resid.plot(ax=ax2)
ax2.set(ylabel='Resid')
decompose_result_mult.seasonal.plot(ax=ax3)
ax3.set(ylabel='Seasonal')
```

### 6.1 Additive Seasonal Decompose with Period 12

```python
decompose_result_mult = seasonal_decompose(analysis, model="additive", extrapolate_trend='freq', period=12)

trend = decompose_result_mult.trend
seasonal = decompose_result_mult.seasonal
residual = decompose_result_mult.resid

fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(15,8))
decompose_result_mult.trend.plot(ax=ax1)
ax1.set(ylabel='Trend')
decompose_result_mult.resid.plot(ax=ax2)
ax2.set(ylabel='Resid')
decompose_result_mult.seasonal.plot(ax=ax3)
ax3.set(ylabel='Seasonal')
```

### 8. Localized Regression and Localized Weighted Regression Smoothing

```python
from statsmodels.nonparametric.smoothers_lowess import lowess

df_original = data.copy()
df_original.set_index('date', inplace=True)
# Moving Average
df_ma = df_original['values'].rolling(3, center=True, closed='both').mean()

# Loess Smoothing (5% and 15%)
df_loess_5 = pd.DataFrame(lowess(df_original['values'].values, np.arange(len(df_original['values'].values)), frac=0.05)[:, 1], index=df_original.index, columns=['values'])
df_loess_15 = pd.DataFrame(lowess(df_original['values'].values, np.arange(len(df_original['values'].values)), frac=0.15)[:, 1], index=df_original.index, columns=['values'])

# Plot
fig, axes = plt.subplots(4,1, figsize=(7, 7), sharex=True, dpi=120)
df_original['values'].plot(ax=axes[0], color='k', title='Original Series')
df_loess_5['values'].plot(ax=axes[1], title='Loess Smoothed 5%')
df_loess_15['values'].plot(ax=axes[2], title='Loess Smoothed 15%')
df_ma.plot(ax=axes[3], title='Moving Average (3)')
fig.suptitle('How to Smoothen a Time Series', y=0.95, fontsize=14)
plt.show()
```
